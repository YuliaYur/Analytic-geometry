```python
%matplotlib notebook

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits import mplot3d
import numpy as np
from math import sqrt
from math import pi
```

```python
def draw_cone_3d():
    ax = plt.axes(projection='3d')

    for angle in np.linspace(0, 2 * pi, 200):
        x = np.cos(angle)
        y = np.sin(angle)
        ax.plot3D([x, x], [y, y], [0, 5], color='b', linewidth=0.5)
        ax.plot3D([x, x], [-y, -y], [0, 5], color='b', linewidth=0.5)
    # ax.set_xlim(0, 10)
    # ax.set_ylim(0, 10)
    # ax.set_zlim(0, 10)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
```

```python
draw_cone_3d()
plt.show()
```
